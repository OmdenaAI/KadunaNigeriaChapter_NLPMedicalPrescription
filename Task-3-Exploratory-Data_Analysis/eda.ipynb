```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "/kaggle/input/leveraging-nlp-omdena-medical-diseases/merge_demo_amos_v3.csv"
df = pd.read_csv(file_path, low_memory=False)

# Basic Information
def basic_info(df):
    print("Dataset Information:\n")
    print(df.info())
    print("\nSummary Statistics:\n")
    print(df.describe(include='all'))

# Missing Values Analysis
def missing_values(df):
    missing = df.isnull().sum()
    missing = missing[missing > 0].sort_values(ascending=False)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=missing.index, y=missing.values, palette='viridis')
    plt.xticks(rotation=90)
    plt.title("Missing Values Count")
    plt.xlabel("Columns")
    plt.ylabel("Count")
    plt.show()

# Categorical Data Distribution
def categorical_distribution(df, column):
    plt.figure(figsize=(10, 5))
    sns.countplot(y=df[column], order=df[column].value_counts().index, palette='coolwarm')
    plt.title(f"Distribution of {column}")
    plt.xlabel("Count")
    plt.ylabel(column)
    plt.show()

# Execution
basic_info(df)
missing_values(df)

# Visualizing categorical distribution of disease_name
categorical_distribution(df, 'disease_name')
```
