```python
import requests
import time
import random
import csv
import datetime
from bs4 import BeautifulSoup
import pandas as pd
from urllib.parse import urlparse
```

```python
# Base URL for CDC Malaria
symptom_url = "https://www.cdc.gov/malaria/symptoms/index.html"

treatment_url = "https://www.cdc.gov/malaria/treatment/index.html"

# Request and parse symtom_url
response = requests.get(symptom_url)
soup = BeautifulSoup(response.content, "html.parser")

#Extract Symptoms
# Find all symptoms listed under the class "level-1"

# Find the specific section for symptoms
symptom_section = soup.find('div', {'data-section': 'cdc_symptoms_symptoms'})

# Find the specific section for treatment
treatment_section = soup.find('div', class_='dfe-section__content')

# Extract symptoms only from the desired section
symptom_list = [
    li.get_text(strip=True)
    for li in symptom_section.find_all('li', class_='level-1')
]

# Extract treatment only from the desired section
treatment_list = [
    li.get_text(strip=True)
    for li in treatment_section.find_all('li', class_='level-1')
]

#Extract diseases name and add into the df
parsed_url = urlparse(symptom_url)
path_parts = parsed_url.path.strip('/').split('/')
name = path_parts[0] if len(path_parts) > 1 else ''

# # Create a DataFrame
df = pd.DataFrame({
    'Name': name,
    'Symptom': pd.Series(symptom_list),
    'Treatment': pd.Series(treatment_list)

})

# # Display the DataFrame
df
```

```python
# Save data to a CSV file
df.to_csv('malaria.csv', index=False)
```
